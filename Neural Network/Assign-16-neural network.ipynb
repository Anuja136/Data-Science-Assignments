```python
df1 =df.drop(columns=['month','day'])
```

```python
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df1['size_category'] = label_encoder.fit_transform(df1['size_category'])
```

```python
df1['size_category'].unique()
```

```python
df1.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
import numpy
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)

X = df1.iloc[:,:-1]             # load pima indians dataset
Y = df1.iloc[:,-1:]             # split into input (X) and output (Y) variables
    
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=28, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(X, Y, validation_split=0.33, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# Visualize training history

history.history.keys()
```

```python
# summarize history for accuracy

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# summarize history for loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
