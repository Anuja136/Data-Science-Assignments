# Hierarchical clustering

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
# load dataset
Crime = pd.read_csv("crime_data.csv")   
Crime
```

```python
Crime.rename({"Unnamed: 0" : "Cities"}, axis = 1 ,inplace = True)
Crime
```

```python
Crime.info()
```

```python
# Normalization function  (x-mean/std)
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
Crime_1 = norm_func(Crime.iloc[:,1:])
Crime_1
```

```python
# create dendrogram
Dendogrm=sch.dendrogram(sch.linkage(Crime_1,method="average"))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'average')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(Crime_1)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
Crime["clusterid"] = Clusters
```

```python
Crime.groupby("clusterid").agg(["mean"]).reset_index()
```

```python
Crime[Crime["clusterid"] == 0] # Here we can see 0 cluster has lowest crime rate compared to others
```

```python
# plot clusters
plt.figure(figsize = (10,7))
plt.scatter(Crime["clusterid"], Crime["Murder"], c = hc.labels_)
```

# K-Means Clustering

```python
# import libraries
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize
from sklearn.preprocessing import StandardScaler
```

```python
Crime = pd.read_csv("crime_data.csv")   # Loading Dataset
Crime
```

```python
Crime.rename({"Unnamed: 0" : "Cities"}, axis = 1 ,inplace = True)
Crime
```

```python
Crime.info()
```

```python
# Normalisation Function 
Crime_1 = StandardScaler().fit_transform(Crime.iloc[:,1:])
```

```python
Crime_1
```

# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 


```python
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(Crime_1)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
clusters4=KMeans(4,random_state=42).fit(Crime_1)
clusters4
```

```python
clusters4.labels_
```

```python
clusters4.cluster_centers_ # This will show centers of each clusters
```

```python
#Assign clusters to the data set
Crime['cluster4id'] = clusters4.labels_
Crime
```

```python
# Group data by Clusters (K=4)
Crime.groupby(["cluster4id"]).agg(["mean"]).reset_index()
```

```python
Crime[Crime['cluster4id']==0]
```

```python
# plot clusters
plt.figure(figsize=(10,7))
plt.scatter(Crime["cluster4id"],Crime["Murder"],c=clusters4.labels_)
```

# DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Crime = pd.read_csv("crime_data.csv")   # Loading Dataset
Crime
```

```python
Crime.info()
```

```python
Crime.drop(['Unnamed: 0'], axis = 1, inplace = True)
Crime
```

```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
Crime_norm=StandardScaler().fit_transform(Crime)
Crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(Crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
Crime['clusters']=dbscan.labels_
Crime
```

```python
Crime.groupby('clusters').agg(['mean']).reset_index()
```

```python
# plot scatter 
plt.figure(figsize = (8, 5))  
plt.scatter(Crime['clusters'],Crime['UrbanPop'], c = dbscan.labels_) 
```

2 - Some correlation is visible, but low murder rates in a city seem to indicate lower number of rapes and vice versa

3 - As with murder and assault, also rates of rape and assault show clearer correlations
